```python
#Create a plot and export out the file for use
def myplot(x,y,style,title,xlabel,ylabel,figsize=(20,5),filename="filenamerf[:4]_refData_NDVI_EVI.png"):
    fig, axes = plt.subplots()
    ax=y.plot(ax=axes,figsize=figsize,title=title,style=style)
    plt.xticks(rotation=90)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # set xaxis major Dateformater
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    # set grid lines
    ax = plt.gca().grid(True)
    fig.savefig(filename,dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab
import numpy as np
from matplotlib.pyplot import *
from scipy.interpolate import *
from numpy import *
%matplotlib inline


def reflectance_flux_data(filenamerf, filenamefs):
    print(filenamerf,filenamefs)
    
    #solar reflectance data
    rf_data = pd.read_csv(filenamerf, na_values=['NA'])
    rf_data['Date'] = pd.to_datetime(rf_data['Date'])
    rf_data = rf_data.set_index(['Date'])
    
    # Flux site data 
    flux_data = pd.read_csv(filenamefs)
    flux_data['date']=pd.to_datetime(flux_data['time'],format='%Y%j')
    flux_data.index = flux_data['date']
    flux_data= flux_data.ix[:,'time':]
    
    #Graph reflectance data for NDVI and EVI
    rf_data.ix[:,'NDVI':'EVI'].plot()
    figsize=(25,15)
    pylab.savefig(filenamerf[:6] + '_reflectance_data.jpeg')
    
    #Graph reflectance data for EVI and LSWI
    rf_data.ix[:,'EVI':'LSWI'].plot()
    figsize=(25,15)
    pylab.savefig(filenamerf[:6] + '_reflectance_data2.jpeg')

    #Graph reflectance data for all three of the vegetation indices NDVI, EVI, and LSWI
    rf_data.ix[:,'NDVI':'LSWI'].plot()
    figsize=(25,15)
    pylab.savefig(filenamerf[:6] + '_reflectance_data3.jpeg')
    
    ##plot data with matplotlib
    #plot the daily max temperature for the flux eddy site and then format 
    #the dates by month, day, and year
    import matplotlib.pyplot as plt
    import matplotlib.dates as mdates
    plt.figure(figsize=(20,5))
    plt.xticks(rotation=80)
    ax =plt.plot(flux_data.date, flux_data['Ta_max (Deg C)'],'g-')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_fluxSite_MaxTempC.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_fluxSite_MaxTempC.jpeg')
    
    #graph the y-axis --> GF_NDVI, with the Date as the x-axis. then print the graph
    plt.figure(figsize=(25,15))
    plt.xticks(rotation=90)
    plt.plot(rf_data.index, rf_data['GF_NDVI'],'b-')
    plt.title('GF NDVI')
    plt.xlabel("Date")
    plt.ylabel("NDVI")
    # set xaxis major Dateformater
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    # set grid lines
    ax1 = plt.gca()
    ax1.grid(True)
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_ref_NDVI_Time.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_ref_NDVI_Time.jpeg')
    
    #graph the y-axis GF_NDVI and GF_EVI with the Date acting as the x-axis.
    plt.figure(figsize=(25,15))
    plt.xticks(rotation=90)
    plt.title("NDVI and EVI over time")
    plt.xlabel("Date")
    plt.ylabel("NDVI and EVI")
    plt.plot(rf_data.index, rf_data['GF_NDVI'])
    plt.plot(rf_data.index, rf_data['GF_EVI'])
    
    #create a subset of data for just the Vegetation Indices and the Mean Daily Temperature Ta_mean_(Deg_C)
    rf_subset= rf_data[['GF_NDVI','GF_EVI']]
    
    #plot NDVI and EVI for the new subset during the years it has relevant data
    plt.figure(figsize=(25,15))
    plt.xticks(rotation=90)
    plt.title("NDVI and EVI over time")
    plt.xlabel("Date")
    plt.ylabel("NDVI and EVI")
    plt.plot(rf_data.index, rf_subset['GF_NDVI'])
    plt.plot(rf_data.index, rf_subset['GF_EVI'],'r-')
    
    #create a new tool to make graphing easier
    def myplot(x,y,style,title,xlabel,ylabel,figsize=(20,5)):
        fig, axes = plt.subplots()
        ax=y.plot(ax=axes,figsize=figsize,title=title,style=style)
    
    #merge the flux site data set to the MODIS reflectance data for all sites
    mdf = rf_subset.merge(flux_data,left_index=True,right_index=True)
    y=mdf[['GF_NDVI','GF_EVI','Ta_min (Deg C)','Ta_max (Deg C)']]
    mdf["Date"]= mdf.index
    myplot(mdf.Date,y,'','Temperature and EVI','Date','Deg C')
    
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_ref_NDVI_EVI_MinMaxTempTime.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_ref_NDVI_EVI_MinMaxTempTime.jpeg'
    #Create chart with multiple y axis for NDVI affected by daily mean temperature
    fig, ax1 = plt.subplots()
    t = mdf.index
    s1 = mdf[['GF_NDVI']]
    ax1.plot(t, s1, 'g--')
    figsize=(40,20)
    ax1.set_xlabel('Date')
    # Make the y-axis label, ticks and tick labels match the line color.
    ax1.set_ylabel('Normalized Difference Vegetation Index(NDVI)', color='g')
    ax1.tick_params('y', colors='g')

    ax2 = ax1.twinx()
    s2 =mdf[['Ta_mean (Deg C)']]
    ax2.plot(t, s2, 'r-')
    figsize=(40,20)
    ax2.set_ylabel('Temp Deg C', color='r')
    ax2.tick_params('y', colors='r')
    # set grid lines
    ax = plt.gca().grid(True)
    
    
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_fluxSite_NDVI_DTemp_Time.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_fluxSite_NDVI_DTemp_Time.jpeg')
    ##############################
    
    #Create a chart with multiple y axis for NDVI and EVI affected by daily mean temperature
    #Create chart with multiple y axis
    fig, ax1 = plt.subplots()
    ax1.plot(subplots=True, figsize=(20, 20));
    t = mdf.index
    s1 = mdf[['GF_NDVI','GF_EVI']]
    ax1.plot(t, s1, 'b--')
    ax1.set_xlabel('Date')
    # Make the y-axis label, ticks and tick labels match the line color.
    ax1.set_ylabel('Vegitation Indices (NDVI and EVI)', color='b')
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.plot(subplots=True, figsize=(10, 10));
    s2 =mdf[['Ta_mean (Deg C)']]
    ax2.plot(t, s2, 'r-')
    ax2.set_ylabel('Temp Deg C', color='r')
    ax2.tick_params('y', colors='r')
    fig.tight_layout()
    
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_fluxSite_NDVI_EVI_DTempTime.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_fluxSite_NDVI_EVI_DTempTime.jpeg')
    #################################
    
    #Create a chart comparing precipitation in mm/day to Temperature in Deg C overtime
    
    y2=mdf[['P_F','Ta_mean (Deg C)']]
    myplot(mdf.Date,y2,'','Temperature and Precipitation','Date','Deg C')
    
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_fluxSite_Precip_DTempTime.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_fluxSite_Precip_DTempTime.jpeg')
    
    #Create a chart that compares incoming solar radiation to the observed Daily Mean Temperature in Deg C
    fig, ax1 = plt.subplots()
    t = mdf.index
    s1 = mdf[['SW_IN_F']]
    ax1.plot(t, s1, 'b--')
    ax1.set_xlabel('Date')
    # Make the y-axis label, ticks and tick labels match the line color.
    ax1.set_ylabel('Incoming Shortwave Radiation, W m-2', color='b')
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    s2 =mdf[['Ta_mean (Deg C)']]
    ax2.plot(t, s2, 'r-')
    ax2.set_ylabel('Temp Deg C', color='r')
    ax2.tick_params('y', colors='r')
    fig.tight_layout()
    #plt.subplots(figsize=(10,15))

    # set grid lines
    ax = plt.gca().grid(True)
    
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_fluxSite_SunRad_DTemp_Time.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_fluxSite_SunRad_DTemp_Time.jpeg')
    
    #Create a chart comparing GPP to NDVI and EVI
    y3=mdf[['GPP_NT_VUT_REF','GF_NDVI','GF_EVI']]
    myplot(mdf.Date,y3,'','Gross Primary Product and Vegetation Indices','Date','g C m-2 day-1')
    
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_fluxSite_NDVI_EVI_GPP_Time.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_fluxSite_NDVI_EVI_GPP_Time.jpeg')
    
    ####
    # Lastly, create a regression model for a scatterplot comparing the GPP to NDVI and EVI
    ####
    
    x=mdf['GPP_NT_VUT_REF']
    y4=mdf['GF_NDVI']
    y5=mdf['GF_EVI']
    
    pl = polyfit(x,y4,1)
    pl2 = polyfit(x,y5,1)
    print(pl)
    print(pl2)
    
    plot(x,y4,'o')
    plot(x,polyval(pl,x),'r-')
    
    plot(x,y5,'o')
    plot(x,polyval(pl2,x),'g-')


    plt.scatter(mdf.GPP_NT_VUT_REF, mdf.GF_NDVI)
    plt.scatter(mdf.GPP_NT_VUT_REF, mdf.GF_EVI)
    plt.xlabel("Gross Primary Product")
    plt.ylabel("Gap Filled NDVI and EVI")
    plt.title("Relationship between GPP and NDVI and EVI ")
    plt.show()
    
    fig = plt.gcf()
    fig.savefig(filenamerf[:6] + '_merg_fluxSite_NDVI_EVI_GPP_regr.jpeg')
    plt.show()
    fig.savefig(filenamerf[:6] + '_merg_fluxSite_NDVI_EVI_GPP_regr.jpeg')
```
